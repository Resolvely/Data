```python
#Notice col ACTMT75 had whitespace
df.rename(mapper = {'ACTMT75   ':'ACTMT75'}, axis = 1, inplace = True)
```

```python
#Handle Outliers of SAT/ACT Test Scores
satact_col = ['SATVR25', 'SATVR50', 'SATVR75', 'SATMT25', 'SATMT50', 'SATMT75',
              'ACTCM25', 'ACTCM50', 'ACTCM75', 'ACTEN25', 'ACTEN50', 'ACTEN75',
              'ACTMT25','ACTMT50', 'ACTMT75']
for col in satact_col:
    df.loc[:, col] = stats.mstats.winsorize(df[col], limits=[0.01, 0.01])
```

```python
#Visualize distribution of SATVR Scores
satvr_df= df.dropna(subset = ['UNITID','SATVR25', 'SATVR50', 'SATVR75'])

satvr_melted = pd.melt(satvr_df, id_vars = ['UNITID'], value_vars = ['SATVR25', 'SATVR50', 'SATVR75'],
                     var_name = 'Section', value_name = 'Score')
```

```python
# Plot the distribution of SATVR scores using Seaborn
plt.figure(figsize=(10, 6))
sns.kdeplot(data=satvr_melted, x='Score', hue='Section', fill=True, common_norm=False, alpha=0.4)
plt.title('Distribution of SAT: Evidence-based Reasoning + Writing Scores')
plt.xlabel('Score')
plt.ylabel('Density')
plt.show()
```
