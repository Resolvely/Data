```python
#Visualize missing values across dataset
col_with_null = df.isnull()

plt.figure(figsize = (10,6))
sns.heatmap(col_with_null, cbar = False, cmap = 'viridis')
plt.title('Missing Data Distribution')
plt.show()
```

```python
df.drop([col for col in df.columns if col.endswith('_P')], axis=1, inplace = True)
```

```python
df.shape
```

```python
nan_count = np.sum(df.isnull(), axis = 0)
nan_count
```

```python
condition = nan_count != 0
nan_col = list(nan_count[condition].index)

nan_col_types = df[nan_col].dtypes
nan_col
```

```python
df.fillna(0, inplace = True) #Possible that missing values indicate that no aid was given
```

```python
col_list = list(df.columns)

for col in col_list:
        df[col] = stats.mstats.winsorize(df[col], limits=[0.01, 0.01])
```
