#Exploring Ethnicity Demographics using a Pie Chart:

```python
#Exploring Ethnicity Demographics using a Pie Chart:
#The following ethinicities include: American/Alaska Indian, Asian, Black/African American, Hispanic/Latin, Native Hawaiian/Pacific Islander, White, Two/More, Unknown 
# Define the columns (selected only 'total' columns)
total_ethnicity_col = ['EFYAIANT', 'EFYASIAT', 'EFYBKAAT', 'EFYHISPT', 'EFYNHPIT', 'EFYWHITT', 'EFY2MORT', 'EFYUNKNT']

# Calculate the total count for each ethnicity
total_counts = {col: df[col].sum() for col in total_ethnicity_col}

# Convert to DataFrame for easier plotting
total_ethnicity_df = pd.DataFrame(list(total_counts.items()), columns=['Ethnicity', 'Total'])

# Extract labels and sizes
labels = total_ethnicity_df['Ethnicity']
sizes = total_ethnicity_df['Total']

# Plot pie chart
plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct= '%1.1f%%')

# Customization
plt.title('Distribution of Ethnicities', fontsize=14)
plt.show()
```
